```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
stock = pd.read_pickle('./stock_price_all.pkl')
```

```python
stock.columns
```

```python
stock_df = stock[['Adj Close_TSLA','Adj Close_GM', 'Adj Close_F','Price_DJAHD','Europe Brent Spot Price FOB  Dollars per Barrel']]
```

```python
stock_df.columns
```

```python
stock_df = stock_df.rename(columns={'Adj Close_TSLA':'Tesla', 'Adj Close_GM':'GM', 'Adj Close_F':'Ford', 'Price_DJAHD':'Dow Jones Automobiles',
       'Europe Brent Spot Price FOB  Dollars per Barrel':'Crude Oil (per Barrel)'})
```

```python
plt.rcParams["figure.figsize"] = (20,10)
stock_df.plot()
```

```python
plt.rcParams["figure.figsize"] = (20,10)
#plt.plot(stock_df)
stock_df.plot()
plt.title("Price Changes",fontsize=24, fontweight='bold')
plt.xticks(stock_df.index[::50], rotation=10, fontsize=12)
plt.yticks(fontsize=12)
plt.ylabel("Price (in dollars)", fontsize=20,fontweight='bold')
plt.xlabel("Date",fontsize=20, fontweight='bold')
handles, labels = plt.gca().get_legend_handles_labels()
order = [3,0,4,1,2]
plt.legend([handles[i] for i in order], [labels[i] for i in order], fontsize=18)
plt.show()
```

```python
sent['Date'] = pd.to_datetime(sent['Date'])
```

```python
sent = sent.set_index('Date')
```

```python
sent.head()
```

```python
sent.columns
```

```python
sent_df_nltk = sent[['summary_nltk_compound','title_nltk_compound','content_nltk_compound']]
```

```python
sent_df_nltk = sent_df_nltk.rename(columns={'summary_nltk_compound':'summary','title_nltk_compound':'title','content_nltk_compound':'content'})
```

```python
sent_df_flair = sent[['summary_flair_score', 'title_flair_score', 'content_flair_score']]
```

```python
sent_df_flair = sent_df_flair.rename(columns={'summary_flair_score':'summary', 'title_flair_score':'title', 'content_flair_score':'content'})
```

```python
plt.rcParams["figure.figsize"] = (20,10)
sent_df_nltk.plot()
plt.title("Tesla Articles' Sentiment Score Changes (NLTK)",fontsize=24, fontweight='bold')
plt.xticks(sent_df_nltk.index[::50], rotation=10, fontsize=12)
plt.yticks(fontsize=12)
plt.ylabel("Sentiment Score", fontsize=20,fontweight='bold')
plt.xlabel("Date",fontsize=20, fontweight='bold')
handles, labels = plt.gca().get_legend_handles_labels()
order = [2,0,1]
plt.legend([handles[i] for i in order], [labels[i] for i in order], fontsize=18)
plt.show()
```

```python
plt.rcParams["figure.figsize"] = (20,10)
sent_df_flair.plot()
plt.title("Tesla Articles' Sentiment Score Changes (Flair)",fontsize=24, fontweight='bold')
plt.xticks(sent_df_nltk.index[::50], rotation=10, fontsize=12)
plt.yticks(fontsize=12)
plt.ylabel("Sentiment Score", fontsize=20,fontweight='bold')
plt.xlabel("Date",fontsize=20, fontweight='bold')
plt.show()
```
